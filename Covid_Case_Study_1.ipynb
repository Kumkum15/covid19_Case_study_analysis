```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#grouping countries and summing up the values

countries_group = data_confirmed.groupby('Country/Region').sum(numeric_only = True)

#Getting the top 5 countries
top_countries = countries_group.iloc[:, -1].sort_values(ascending=False).head(5)

#plotting the graph
plt.figure(figsize=(12,6))
plt.bar(top_countries.index, top_countries.values, color='coral')
plt.title("Confirmed cases for top countries", fontsize=16)
plt.xlabel('Top Countries', fontsize = 12)
plt.ylabel('Confirmed cases', fontsize = 12)
plt.xticks(rotation=45)

plt.show()
```

```python
#Time series graph for china

Data_Of_China = data_confirmed[data_confirmed['Country/Region']=='China']

#Confirmed cases sum
china_confirmed_cases = Data_Of_China.iloc[:,4:].sum()

#Changing the datatype of dates from string to datetime
china_confirmed_cases.index = pd.to_datetime(china_confirmed_cases.index)

plt.figure(figsize=(12,5))
plt.plot(china_confirmed_cases, color='green', linewidth = 2)
plt.title("Covid cases in China")
plt.xlabel='Dates'
plt.ylabel='No. of Confirmed cases'
plt.xticks(rotation=45)

plt.show()
```

```python
data_confirmed['Province/State'] = data_confirmed['Province/State'].fillna("All Provinces")
data_confirmed['Province/State']
```

```python
required_countries = ["Germany", "France", "Italy"]
data_confirmed = data_confirmed[data_confirmed['Country/Region'].isin(required_countries)]
#melted_data = data_confirmed.melt(data_confirmed, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Confirmed')
melted_data = data_confirmed.melt(
    id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
    var_name='Date',
    value_name='Confirmed'
)
melted_data['Date'] = pd.to_datetime(melted_data['Date'], format = '%m/%d/%y')
melted_data=melted_data.sort_values(by=['Country/Region', 'Date'])
melted_data['Daily_New_Cases'] = melted_data.groupby('Country/Region')['Confirmed'].diff().fillna(0)
melted_data['Daily_New_Cases'] = melted_data['Daily_New_Cases'].apply(lambda x: max(0, x))

peak_surges=[]

for country in required_countries:
    country_data = melted_data[melted_data['Country/Region'] == country]
    if not country_data.empty:
        # Find the row with the maximum daily new cases for the current country
        peak_row = country_data.loc[country_data['Daily_New_Cases'].idxmax()]
        peak_date = peak_row['Date'].strftime('%Y-%m-%d')
        peak_cases = int(peak_row['Daily_New_Cases'])
        peak_surges.append({'Country': country, 'Date': peak_date, 'Cases': peak_cases})
        print(f"{country}: Peak daily surge of {peak_cases:,} cases on {peak_date}")
    else:
        print(f"No data found for {country}")

# Find the overall highest single-day surge
if peak_surges:
    highest_surge = max(peak_surges, key=lambda item: item['Cases'])
    print(f"The country that experienced the highest single-day surge is {highest_surge['Country']} with {highest_surge['Cases']:,} new cases, which occurred on {highest_surge['Date']}.")
else:
    print("Could not determine overall highest surge due to data issues.")
```
